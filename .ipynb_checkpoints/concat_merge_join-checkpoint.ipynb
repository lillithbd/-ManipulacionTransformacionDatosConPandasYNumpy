### Importando librería

```python
import pandas as pd
import numpy as np
```

### Crear nuevos DataFrames

```python
df1 = pd.DataFrame({'A':['A0', 'A1', 'A2','A3'],
                    'B':['B0', 'B1', 'B2','B3'],
                	'C':['C0', 'C1', 'C2','C3'],
                	'D':['D0', 'D1', 'D2','D3']})
df1
```

```python
df2 = pd.DataFrame({'A':['A4', 'A5', 'A6','A7'],
                	'B':['B4', 'B5', 'B6','B7'],
                	'C':['C4', 'C5', 'C6','C7'],
                	'D':['D4', 'D5', 'D6','D7']})
df2
```

### Concatenar los DataFrames

```python
pd.concat([df1,df2])
```

### Corrigiendo índices

```python
pd.concat([df1,df2], ignore_index=True)
```

### Ampliar de manera horizontal la concatenación

```python
pd.concat([df1,df2], axis = 1)
```

### Merge

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
                     'A' : ['A0', 'A1', 'A2','A3'],
                     'B': ['B0', 'B1', 'B2','B3']})
izq
```

```python
der = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
                      'C' : ['C0', 'C1', 'C2','C3'],
                       'D': ['D0', 'D1', 'D2','D3']})
der
```

### Merge de izquierda a derecha 
##### el de la izquierda es el primero, el de la derecha es el que se encuentra ubicado dentro de los paréntesis

```python
izq.merge(der, on='key')
```

### Merge 2

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
                     'A' : ['A0', 'A1', 'A2','A3'],
                     'B': ['B0', 'B1', 'B2','B3']})
izq
```

```python
der = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2','k3'],
                      'C' : ['C0', 'C1', 'C2','C3'],
                       'D': ['D0', 'D1', 'D2','D3']})
der
```

```python
izq.merge(der, left_on = 'key', right_on='key_2')
```

### Merge 3

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
                     'A' : ['A0', 'A1', 'A2','A3'],
                     'B': ['B0', 'B1', 'B2','B3']})
izq
```

```python
der = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2',np.nan],
                     'C' : ['C0', 'C1', 'C2','C3'],
                     'D': ['D0', 'D1', 'D2','D3']})
der
```

```python
izq.merge(der, left_on = 'key', right_on='key_2')
```

#### Si tenemos un NaN en nuestro DataFrame, pandas no lo detectará como un match. Se soluciona con How, dando así, una preferencia.

```python
izq.merge(der, left_on = 'key', right_on='key_2', how='left')
```

```python
izq.merge(der, left_on = 'key', right_on='key_2', how='right')
```

### Join - Index Match

```python
izq = pd.DataFrame({'A' : ['A0', 'A1', 'A2'],
                    'B': ['B0', 'B1', 'B2']},
                    index =['k0', 'k1', 'k2'])

der = pd.DataFrame({'C' : ['C0', 'C1', 'C2'],
                    'D': ['D0', 'D1', 'D2']},
                    index =['k0', 'k2', 'k3'])
```

```python
izq.join(der)
```

```python
izq.join(der, how='inner')
```

```python
izq.join(der, how='left')
```

```python
izq.join(der, how='right')
```

```python
izq.join(der, how='outer')
```
