### Importando Librerías

```python
import pandas as pd
```

### Leer archivo

##### sep = separador. Por default viene en comas, pero esto se puede cambiar
##### header(0) = toma la primera fila como el header. 
##### header(None) = cuando el archivo no tenga header
##### names = Para especificar el nombre de los encabezados

```python
df_books = pd.read_csv('bestsellers-with-categories.csv', sep=',', header=0, names = ['Name', 'Author', 'User Rating', 'Reviews', 'Price', 'Year', 'Genre'])
```

```python
df_books
```

#### Filtrando datos

```python
df_books[0:4]
```

#### Filtrar por eje de columnas

```python
df_books['Name']
```

```python
df_books[['Name', 'Author']]
```

```python
df_books[['Name', 'Author', 'Year']]
```

#### Filtrando con Loc (Labels)

```python
df_books.loc[:]
```

```python
df_books.loc[0:4]
```

```python
df_books.loc[0:4, ['Name', 'Author']]
```

```python
df_books.loc[:, ['Reviews']] * -1
```

```python
df_books.loc[:, ['Author']] == 'JJ Smith'
```

#### Filtrando con iloc (Índice)

```python
df_books.iloc[:]
```

```python
df_books.iloc[:,0:3]
```

```python
df_books.iloc[1,3] * -1
```

```python
df_books.iloc[:2, 2:]
```
