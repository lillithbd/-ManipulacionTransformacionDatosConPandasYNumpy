### Importando Librerías

```python
import pandas as pd
```

### Cargar Datos

```python
df_books = pd.read_csv('bestsellers-with-categories.csv',sep=',',header=0)
df_books.head(2)
```

### Creamos una función

```python
def two_times(value):
    return value * 2
```

### Aplicamos funcion a columna User Rating

```python
df_books['User Rating'].apply(two_times)
```

### Crear nueva columna calculada

```python
df_books['Rating_2'] = df_books['User Rating'].apply(two_times)
df_books
```

### Crear nueva columna calculada con función lambda

```python
df_books['Rating_2'] = df_books['User Rating'].apply(lambda x : x * 3)
df_books
```

### Usando distintas columnas para hacer operaciones

```python
df_books['Rating_2'] = df_books.apply(lambda x : x['User Rating'] * 2 if x['Genre'] == 'Fiction' else x['User Rating'], axis = 1)
```
